# Intorduction

The gravitation appears deceptively simple when it is put into mathematical form.
The weakest among the lot! Its dependence on mass and the infinite range of effect
creates an insurmountable difficulty when we consider the dynamics of any celestial
body. In common dictum, each body attracts the other, so the motion of a body has
a very complicated dependence on configurations of other bodies. N-body problem
is about this. The aim of this protocol is to consider an assembly of N-bodies and
to study the cause and effect among themselves.

From classical mechanics gives a few handful way of looking at the problem of classical dynamics. Lagrangian and Hamiltonian point of view is essentially similar to each other, but preferences of one over the other depend on the nature of the problem. Hamiltonian method opens a new way of visualizing the problem in the phase space. Since, N-body problem consists of large number of particles, so Hamiltonian point of view is more feasible. \\
The motion of N point masses in their mutual gravitational field is the classical N -body problem.
The Hamiltonian of a system is 
\begin{equation}
    H  = \sum_{i=1}^{N} \frac{\boldsymbol{{p}}^2_i}{2m_i}-\sum_{i=1}^{N}\sum_{j=i+1}^{N}\frac{G m_i m_j}{|\boldsymbol{q}_i-\boldsymbol{q}_j|}
\end{equation}
With canonical momentum $\boldsymbol{p}_i=mv_i$.
$m_i$ is the mass of particle located at $r_i$.
\begin{gather}\label{dynamical}
    \boldsymbol{q}_i=\frac{\partial H}{\partial \boldsymbol{p}_i}
    \implies \frac{d\boldsymbol{r}_i}{dt}=\boldsymbol{v}_i\\
    \boldsymbol{p}_i=-\frac{\partial H}{\partial \boldsymbol{q}_i}
    \implies \frac{d\boldsymbol{v}_i}{dt}=\boldsymbol{a}_i
\end{gather}
Where $a_i$ is the acceleration,
\begin{equation}\label{acceleration}
    \boldsymbol{a}_i(t)=\sum_{j\neq i}^{N} =Gm_j\frac{\boldsymbol{r}_{ij}}{r^3_{ij}}
\end{equation}
Where 
\begin{gather}\label{absr}
    \boldsymbol{r}_{ij}=\boldsymbol{r}_j(t)-\boldsymbol{r}_i(t)\\
    r_{ij}= |\boldsymbol{r}_{ij}|
\end{gather}
Another formula that will be required further is the jerk
\begin{equation}\label{jerk}
    \boldsymbol{\dot{a}}_i(t)=\sum_{j\neq i}^{N} Gm_j \left(\frac{\boldsymbol{v}_{ij}}{r^3_{ij}}-\frac{3(\boldsymbol{v}_{ij}\cdot\boldsymbol{r}_{ij})}{r_{ij}}\boldsymbol{r}_{ij} \right)
\end{equation}
$\boldsymbol{v}$ is velocity of $i$th particle.

# A step Towards Calculation


The aim of dynamics is to find the position of a particle ($\boldsymbol{r}_i$) for a given time $t$. To achieve this, there is need to integrate the differential equations (\ref{dynamical}). The expression for $a$ from eq (\ref{acceleration}) will be put into equations (\ref{dynamical}). This equation will be solved numerically. For numerical calculation, several numerical integrators will be used (as required in Exercise -1).
Let's consider a differential equation
\begin{equation}
    \frac{dy}{dt}=f(t,y)
\end{equation}
From the definition of derivative 
\begin{equation}
   \lim_{\Delta t \to 0} \frac{y(t+\Delta t)-y(t)}{\Delta t}= f(t,y)
\end{equation}
For simplicity we take $\Delta t =h$
For a finite interval 
$$\frac{y(t+\Delta t)-y(t)}{\Delta t}= f(t,y)$$
0r,
\begin{equation}\label{basic}
    \frac{y(t+h)-y(t)}{h}= f(t,y)
\end{equation}
Integration will be performed from some $t=t_{initial}$ to $t=t_{final}$. The interval between $t_i$ to $t_f$ has been divided into N small part of width $h$. So $t_n=n \Delta t = nh$\\
So equation (\ref{basic}) could be written in the following form
\begin{gather}\label{numeric1}
    y_{n+1}-y_n=hf 
    \implies y_{n+1}=y_n+f(n,y_n)
\end{gather}
This is the simplest form of integrating an equation numerically, and is called the Euler Method. 
From equation (3) and equation (\ref{numeric1}), it follows,
\begin{gather}\label{euler}
    v_{n+1}=v_n+a_n\Delta t\\
    r_{n+1}=r_n+v_n\Delta t
\end{gather}

# Leap Frog Method 


 The leap frog method is an improved version of the Euler method. In this scheme, velocity is calculated at $t_n+\Delta t/2$ instead of $t_n$ or $t_{n+1}$. Calculation consists of following steps.\\
 Step-I\\
 $$r_{1/2}=r_0+v_0\frac{\Delta t}{2}$$
 $$a_{1/2}=a(t_{1/2},r_{1/2})$$
 Step-II\\
 $$v_{n+1}=v_n+a_{n+1/2}\Delta t$$
 $$r_{n+3/2}=r_{n+1/2}+v_{n+1} \Delta t$$
 Step -III\\
 $$r_{n+1}=r_{n+1/2}+v_{n+1}\frac{\Delta t}{2}$$
 

# The Verlet method


 This method is similar to Leap frog but with some modifications. So for the calculation purposes we  take following form, 
 \begin{gather}
     \bar v=v_n+1/2(\Delta t a_n)\\
     r_{n+1}=r_n+\Delta t \bar v\\
     v_{n+1}=\bar v+\frac{1}{2}a_{n+1}(\Delta t)
 \end{gather}
 The advantage of this velocity Verlet-method is the higher accuracy in the calculation of the
locations. The Verlet-method is identical to the Leap-frog-method. Note, that also in this
method, only one evaluation of the force term is required per time step.

# Runge-Kutta Method


Eulerâ€™s method is classified as a first-order method because its cumulative truncation
error behaves as O(h). Its basis was the truncated Taylor series,
$$y(t+h)=y(t)+y'(t)h$$
The accuracy of numerical integration can be greatly improved by keeping more
terms of the series. Thus, a nth order method would use the truncated Taylor series
\begin{equation}\label{talor}
    y(t+h)=y(t)+y'(t)h+.......\frac{1}{n!}y^{(n)}(t)h^n
\end{equation}
But now we must derive the expressions for $y'$ , $y''$ , . . . $y^{(n)}$ by repeatedly differentiating $y'= F(t, y)$ and write subroutines to evaluate them. This extra work can be
avoided by using Runge-Kutta methods that are also based on truncated Taylor series, but do not require computation of higher derivatives of $y(t)$. Since, derivation is long, and it would a complete repetition of derivation in protocol booklet, here formula for fourth order Runge -Kutta method have used directly 
The general form of an explicit RK method is
\begin{equation}
    y_{n+1}=y_n+h\sum_i b_i K_i
\end{equation}
where to $K_i$ denote the substeps and $b_i$ are some weights.
For a system of equation
$$\boldsymbol{y}'=\boldsymbol{f}(t,y(t)), \boldsymbol{y}(t_0)=y_0$$
For the fourth order Runge Kutta (RK4) $\boldsymbol{K}'s $ are following forms
\begin{gather}
    \boldsymbol{K}_1=\boldsymbol{f}(t_n,y_n)\\
    \boldsymbol{K}_2=\boldsymbol{f}(t_n+\frac{1}{2}h,y_n+\frac{h}{2}\boldsymbol{K}_1)\\
    \boldsymbol{K}_3=\boldsymbol{f}(t_n+\frac{1}{2}h,y_n+\frac{h}{2}\boldsymbol{K}_2)\\
    \boldsymbol{K}_4=\boldsymbol{f}(t_{n+1},y_n+h\boldsymbol{K}_3)
\end{gather}
So final form of RK4 scheme is following 
\begin{equation}
    \boldsymbol{y}_{n+1}=\boldsymbol{y}_{n}+\frac{h}{6}(\boldsymbol{K}_1+2\boldsymbol{K}_2+2\boldsymbol{K}_3+\boldsymbol{K}_4)
\end{equation}


The error of the scheme is of the order O($h^5$ ). The price for this high accuracy is the required
calculation of the function, which is four times per time step.\\

```python
import numpy as np
import matplotlib.pyplot as plt
import sys



### basic data
# number of bodies
N = 2
# universal gravity constant
G = 1.0

x = np.zeros(N)
y = np.zeros(N)
vx = np.zeros(N)
vy = np.zeros(N)
m = np.zeros(N)
ax = np.zeros(N)
ay = np.zeros(N)

P = 2*np.pi
# set masses to 1.0 and 10**(-3)
# Exercise 2.0
m[1] = 1e-3
m[0] = 1.0-m[1]
sma = 1.0
ecc = 0.5
t_begin = 0.0
t_end = 10*P

fixed_dt = P/500


# what do we need?
# function for acceleration
# functions for euler, euler_cromer, verlet, rk4
# function for adaptive timestep
# function to determine initial positions


def calc_iv(m, sma, ecc):
    x[1] = sma*(1+ecc)
    #x[2]=1
    vy[1] = np.sqrt(G*np.sum(m)/sma * ((1-ecc)/(1+ecc)))



def calc_conserved_quantities(x, y, vx, vy, m):
    E = 0
    for i in range(len(x)):
        # kinetic part of the energy
        E += 0.5*m[i]*(vx[i]**2 + vy[i]**2)
        for j in range(i+1, len(x)):
            r = np.sqrt((x[i]-x[j])**2 + (y[i]-y[j])**2)
            E -= G*m[i]*m[j]/r
    return E



def calc_acceleration(x, y, m):
    for i in range(len(x)):
        ax[i] = 0.0
        ay[i] = 0.0
        for j in range(len(x)):
            if (i == j):
                continue
            r = np.sqrt((x[i] - x[j])**2 + (y[i] - y[j])**2)
            ax[i] += G*m[j] * (x[j] - x[i])/r**3
            ay[i] -= G*m[j] * (y[i] - y[j])/r**3
    return ax, ay
def jerk_a(x, y, vx, vy, m):
    for i in range(len(x)):
        dax[i]=0.0
        day[i]=0.0
        for j in range(len(x)):
            if (i == j):
                continue
            r = np.sqrt((x[i] - x[j])**2 + (y[i] - y[j])**2)
            print(r)
            dax[i] += G*m[j]*((vx[j]-vx[i])/r**3-3*((vx[j]-vx[i])*(x[j] - x[i]))*(x[j] - x[i])/r**5)
            day[i] += G*m[j]*((vy[j]-vy[i])/r**3-3*((vy[j]-vy[i])*(y[j] - y[i]))*(y[j] - y[i])/r**5)
    return dax, day

def calc_timestep():
    #dax, day= jerk_a(x,y,vx,vy,m)
    #d_a= m.sqrt(dax^2+day^2)
    dt = fixed_dt
    return dt

def euler(x, y, vx, vy, dt):
    ax, ay = calc_acceleration(x, y, m)
    x += dt*vx
    y += dt*vy
    vx += ax*dt
    vy += ay*dt

def symplectic_euler(x, y, vx, vy, dt):
    ax, ay = calc_acceleration(x, y, m)
    vx += ax*dt
    vy += ay*dt
    x += dt*vx
    y += dt*vy

def rk4(x, y, vx, vy, dt):
    x0 = x
    y0 = y 
    v0x = vx
    v0y = vy
    a0x, a0y = calc_acceleration(x0, y0, m)

    v1x = v0x 
    v1y = v0y 
    K1x = dt* v1x
    K1y = dt* v1y
    K1vx = dt*a0x
    K1vy = dt*a0y

    a1x, a1y = calc_acceleration(x0 + K1x/2, y0 + K1y/2, m)

    v2x = v0x + 0.5*dt*a1x
    v2y = v0y + 0.5*dt*a1y
    K2x = dt* v2x
    K2y = dt* v2y
    K2vx = dt*a1x
    K2vy = dt*a1y
    a2x, a2y = calc_acceleration(x0 + K2x/2, y0 + K2y/2, m)


    v3x = v0x + 0.5*dt*a2x
    v3y = v0y + 0.5*dt*a2y
    K3x = dt* v3x
    K3y = dt* v3y
    K3vx = dt*a2x
    K3vy = dt*a2y
    a3x, a3y = calc_acceleration(x0 + K3x, y0 + K3y, m)

    v4x = v0x + dt*a3x
    v4y = v0y + dt*a3y
    K4x = dt* v4x
    K4y = dt* v4y
    K4vx = dt*a3x
    K4vy = dt*a3y

    x += (1/6.0)*(K1x + 2*K2x + 2*K3x + K4x)
    y += (1/6.0)*(K1y + 2*K2y + 2*K3y + K4y)
    vx += (1/6.0)*(K1vx + 2*K2vx + 2*K3vx + K4vx)
    vy += (1/6.0)*(K1vy + 2*K2vy + 2*K3vy + K4vy)

def verlet(x, y, vx, vy, dt):
    #calculate a, v tilda, x,y
    ax, ay = calc_acceleration(x, y, m)

    Vtx = 0.5*ax*dt
    Vty = 0.5*ay*dt

    x += dt*vx +Vtx*dt
    y += dt*vy +Vty*dt

    #calculate a(t +dt), vx, vy
    ax,ay = calc_acceleration(x, y, m)
    vx += Vtx + 0.5*ax*dt
    vy += Vty + 0.5*ay*dt

def Leap_frog(x, y, vx, vy, dt):
    ax, ay = calc_acceleration(x, y,m)
    vx +=ax*dt
    vy += ay*dt
    x += dt*(vx+ax*dt)
    y += dt*(vy+ay*dt)


# function to move some bodies to the center of mass system
def move_to_COM(x, y, vx, vy, m):
    xcm = ycm = vxcm = vycm = 0.0
    for i in range(len(x)):
        xcm += m[i]*x[i]
        ycm += m[i]*y[i]
        vxcm += m[i]*vx[i]
        vycm += m[i]*vy[i]

    total_mass = np.sum(m)
    xcm /= total_mass
    ycm /= total_mass
    vxcm /= total_mass
    vycm /= total_mass

    print("Center of mass is (%lf, %lf)" % (xcm, ycm))
    x -= xcm
    y -= ycm
    vx -= vxcm
    vy -= vycm

def cons_2(xp, yp, vxp, vyp):
    s = np.zeros((len(xp), 3, N)) 
    vel = np.zeros((len(xp), 3, N))
    for i in range(N):
        for k in range(len(s)):
            s[k, 0, i] = xp[k][i]
            s[k, 1, i] = yp[k][i]
            vel[k, 0, i] = vxp[k][i]
            vel[k, 1, i] = vyp[k][i]
    for i in range(N):
        for j in range(i+1,N):
            if (i!=0 and j!=0):
                continue;
        
            R = s[:,:,i]-s[:,:,j]
            V = vel[:,:,i]-vel[:,:,j]

            #specific angular momentum
            jj = np.cross(R,V) #vector
            jz = np.sqrt(jj[:,0]**2+jj[:,1]**2+jj[:,2]**2)
            
            #eccentricity
            rr = np.sqrt(R[:,0]**2+R[:,1]**2+R[:,2]**2)
            A = np.cross(V,jj)/(G*sum(m))
            B = np.zeros((len(R),3))
            for k in range (len(R)):
                B[k]= R[k]/rr[k]
            C = A-B #Runge Lenz Vector
            ecc = np.sqrt(C[:,0]**2+C[:,1]**2+C[:,2]**2) 

            #semi-major axis
            a_maj = (jz**2/(G*sum(m)))/(1-ecc**2) 
    return jz, ecc, a_maj
# main program starts here

t = t_begin
calc_iv(m, sma, ecc)
move_to_COM(x, y, vx, vy, m)

fig, axs = plt.subplots()

xp = []
yp = []
Ep = []
tp = []
vxp=[]
vyp=[]
rp = []
#ang_j=[]
#Ecc=[]
#sma= []

while t < t_end:
    dt = calc_timestep()
    #euler(x, y, vx, vy, dt)
    #symplectic_euler(x, y, vx, vy, dt)
    rk4(x, y, vx, vy, dt)
    #Leap_frog(x,y,vx,vy,dt)
    #verlet(x, y, vx, vy, dt)
    E = calc_conserved_quantities(x, y, vx, vy, m)
    xp.append(list(x))
    yp.append(list(y))
    vxp.append(list(vx))
    vyp.append(list(vy))
    Ep.append(E)
    tp.append(t)
    t += dt
    print("t is currently %g" % t, end='\r')

jz,ecc,a_maj = cons_2(xp, yp, vxp, vyp)
print(jz)
Ecc=np.array(ecc)
#ang_j=np.array(jz)
#sma=(a_maj)
#print(type(Ecc))
#print(ecc)
#axs.scatter(xp, yp, s=1.0, c='r')
axs.scatter(tp, Ep, s=1.0, c='b')
axs.set_xlabel("Time in code units")
axs.set_ylabel("Total Energy")
#axs.set_aspect('equal')
#fig.savefig("EX2__1_cc.png")
plt.title('Two-Body Problem : Rk4 Method')
plt.savefig("EX2_1cc.png")
#print(x)
print("Done")
```

```python
#plotting trjactory
import numpy as np
import matplotlib.pyplot as plt
import sys

# this python script originates from
# the live coding session on 7 October 2020
# within the labwork
# author Christoph Schaefer


### basic data
# number of bodies
N = 2
# universal gravity constant
G = 1.0

x = np.zeros(N)
y = np.zeros(N)
vx = np.zeros(N)
vy = np.zeros(N)
m = np.zeros(N)
ax = np.zeros(N)
ay = np.zeros(N)

P = 2*np.pi
# set masses to 1.0 and 10**(-3)
# Exercise 2.0
m[1] = 1e-3
m[0] = 1.0-m[1]
sma = 1.0
ecc = 0.5
t_begin = 0.0
t_end = 10*P

fixed_dt = P/500


# what do we need?
# function for acceleration
# functions for euler, euler_cromer, verlet, rk4
# function for adaptive timestep
# function to determine initial positions


def calc_iv(m, sma, ecc):
    x[1] = sma*(1+ecc)
    #x[2]=1
    vy[1] = np.sqrt(G*np.sum(m)/sma * ((1-ecc)/(1+ecc)))



def calc_conserved_quantities(x, y, vx, vy, m):
    E = 0
    for i in range(len(x)):
        # kinetic part of the energy
        E += 0.5*m[i]*(vx[i]**2 + vy[i]**2)
        for j in range(i+1, len(x)):
            r = np.sqrt((x[i]-x[j])**2 + (y[i]-y[j])**2)
            E -= G*m[i]*m[j]/r
    return E



def calc_acceleration(x, y, m):
    for i in range(len(x)):
        ax[i] = 0.0
        ay[i] = 0.0
        for j in range(len(x)):
            if (i == j):
                continue
            r = np.sqrt((x[i] - x[j])**2 + (y[i] - y[j])**2)
            ax[i] += G*m[j] * (x[j] - x[i])/r**3
            ay[i] -= G*m[j] * (y[i] - y[j])/r**3
    return ax, ay
def jerk_a(x, y, vx, vy, m):
    for i in range(len(x)):
        dax[i]=0.0
        day[i]=0.0
        for j in range(len(x)):
            if (i == j):
                continue
            r = np.sqrt((x[i] - x[j])**2 + (y[i] - y[j])**2)
            print(r)
            dax[i] += G*m[j]*((vx[j]-vx[i])/r**3-3*((vx[j]-vx[i])*(x[j] - x[i]))*(x[j] - x[i])/r**5)
            day[i] += G*m[j]*((vy[j]-vy[i])/r**3-3*((vy[j]-vy[i])*(y[j] - y[i]))*(y[j] - y[i])/r**5)
    return dax, day

def calc_timestep():
    #dax, day= jerk_a(x,y,vx,vy,m)
    #d_a= m.sqrt(dax^2+day^2)
    dt = fixed_dt
    return dt

def euler(x, y, vx, vy, dt):
    ax, ay = calc_acceleration(x, y, m)
    x += dt*vx
    y += dt*vy
    vx += ax*dt
    vy += ay*dt

def symplectic_euler(x, y, vx, vy, dt):
    ax, ay = calc_acceleration(x, y, m)
    vx += ax*dt
    vy += ay*dt
    x += dt*vx
    y += dt*vy

def rk4(x, y, vx, vy, dt):
    x0 = x
    y0 = y 
    v0x = vx
    v0y = vy
    a0x, a0y = calc_acceleration(x0, y0, m)

    v1x = v0x 
    v1y = v0y 
    K1x = dt* v1x
    K1y = dt* v1y
    K1vx = dt*a0x
    K1vy = dt*a0y

    a1x, a1y = calc_acceleration(x0 + K1x/2, y0 + K1y/2, m)

    v2x = v0x + 0.5*dt*a1x
    v2y = v0y + 0.5*dt*a1y
    K2x = dt* v2x
    K2y = dt* v2y
    K2vx = dt*a1x
    K2vy = dt*a1y
    a2x, a2y = calc_acceleration(x0 + K2x/2, y0 + K2y/2, m)


    v3x = v0x + 0.5*dt*a2x
    v3y = v0y + 0.5*dt*a2y
    K3x = dt* v3x
    K3y = dt* v3y
    K3vx = dt*a2x
    K3vy = dt*a2y
    a3x, a3y = calc_acceleration(x0 + K3x, y0 + K3y, m)

    v4x = v0x + dt*a3x
    v4y = v0y + dt*a3y
    K4x = dt* v4x
    K4y = dt* v4y
    K4vx = dt*a3x
    K4vy = dt*a3y

    x += (1/6.0)*(K1x + 2*K2x + 2*K3x + K4x)
    y += (1/6.0)*(K1y + 2*K2y + 2*K3y + K4y)
    vx += (1/6.0)*(K1vx + 2*K2vx + 2*K3vx + K4vx)
    vy += (1/6.0)*(K1vy + 2*K2vy + 2*K3vy + K4vy)

def verlet(x, y, vx, vy, dt):
    #calculate a, v tilda, x,y
    ax, ay = calc_acceleration(x, y, m)

    Vtx = 0.5*ax*dt
    Vty = 0.5*ay*dt

    x += dt*vx +Vtx*dt
    y += dt*vy +Vty*dt

    #calculate a(t +dt), vx, vy
    ax,ay = calc_acceleration(x, y, m)
    vx += Vtx + 0.5*ax*dt
    vy += Vty + 0.5*ay*dt

def Leap_frog(x, y, vx, vy, dt):
    ax, ay = calc_acceleration(x, y,m)
    vx +=ax*dt
    vy += ay*dt
    x += dt*(vx+ax*dt)
    y += dt*(vy+ay*dt)


# function to move some bodies to the center of mass system
def move_to_COM(x, y, vx, vy, m):
    xcm = ycm = vxcm = vycm = 0.0
    for i in range(len(x)):
        xcm += m[i]*x[i]
        ycm += m[i]*y[i]
        vxcm += m[i]*vx[i]
        vycm += m[i]*vy[i]

    total_mass = np.sum(m)
    xcm /= total_mass
    ycm /= total_mass
    vxcm /= total_mass
    vycm /= total_mass

    print("Center of mass is (%lf, %lf)" % (xcm, ycm))
    x -= xcm
    y -= ycm
    vx -= vxcm
    vy -= vycm

def cons_2(xp, yp, vxp, vyp):
    s = np.zeros((len(xp), 3, N)) 
    v = np.zeros((len(xp), 3, N))
    for i in range(N):
        for k in range(len(s)):
            s[k, 0, i] = xp[k][i]
            s[k, 1, i] = yp[k][i]
            v[k, 0, i] = vxp[k][i]
            v[k, 1, i] = vyp[k][i]
    for i in range(N):
        for j in range(i+1,N):
            if (i!=0 and j!=0):
                continue;
        
            R = s[:,:,i]-s[:,:,j]
            V = v[:,:,i]-v[:,:,j]

            #specific angular momentum
            jj = np.cross(R,V) #vector
            jz = np.sqrt(jj[:,0]**2+jj[:,1]**2+jj[:,2]**2)
            
            #eccentricity
            rr = np.sqrt(R[:,0]**2+R[:,1]**2+R[:,2]**2)
            A = np.cross(V,jj)/(G*sum(m))
            B = np.zeros((len(R),3))
            for k in range (len(R)):
                B[k]= R[k]/rr[k]
            C = A-B #Runge Lenz Vector
            ecc = np.sqrt(C[:,0]**2+C[:,1]**2+C[:,2]**2) 

            #semi-major axis
            a_maj = (jz**2/(G*sum(m)))/(1-ecc**2) 
    return jz, ecc, a_maj
# main program starts here

t = t_begin
calc_iv(m, sma, ecc)
move_to_COM(x, y, vx, vy, m)

fig, axs = plt.subplots()

xp = []
yp = []
Ep = []
tp = []
vxp=[]
vyp=[]
rp = []
#ang_j=[]
#Ecc=[]
#sma= []

while t < t_end:
    dt = calc_timestep()
    #euler(x, y, vx, vy, dt)
    #symplectic_euler(x, y, vx, vy, dt)
    rk4(x, y, vx, vy, dt)
    #Leap_frog(x,y,vx,vy,dt)
    #verlet(x, y, vx, vy, dt)
    E = calc_conserved_quantities(x, y, vx, vy, m)
    xp.append(list(x))
    yp.append(list(y))
    vxp.append(list(vx))
    vyp.append(list(vy))
    Ep.append(E)
    tp.append(t)
    t += dt
    print("t is currently %g" % t, end='\r')

jz,ecc,a_maj = cons_2(xp, yp, vxp, vyp)
print(jz)
Ecc=np.array(ecc)
plt.plot(xp,yp)
plt.xlabel("Time in code units")
plt.ylabel("trajactory")
plt.title('trajactory: Rk4 method')
plt.savefig("ex2_2cc.png")
```
